In this presentation I will go through some of the theoretical concepts associated with equality constrained minimization and then demonstrate some practical examples.

In general, an ECM problem has the form:

$$\text{minimize } f(x)$$
$$\text{subject to } Ax = b$$

For the purposes of this problem, we assume that $f$ is convex and twice continuously differentiable with $A \in \mathbb{R}^{pxn}$ and $rank A = p < n$. We also assume that the problem is solvable.

$\textbf{Optimality Condition For ECM Problem:}$ $x^*$ is optimal iff there exists some $v^*$ such that the following holds:

$$Ax^* = b$$
$$\nabla f(x^*) + A^T v^* = 0$$

$\textbf{The Quadratic ECM Problem}$

Similarly to the unconstrained minimization problem, we can formulate a version of the ECM problem that is a convex quadratic. We see that this problem is:

$$\text{minimize } f(x) = \frac12 x^TPx + q^T x + r$$
$$\text{subject to } Ax = b$$

Since we now have a more specific $f$, we can come up with a more specific optimality condition for this problem. We see that in this case $\nabla f(x^*) = Px^* + q$. Thus, the optimality condition becomes:

$$Ax^* = b$$
$$Px^* + q + A^T v^* = 0$$

We can refactor these two equations into the following matrix system:

$$\begin{bmatrix} P & A^T \\ A & 0 \end{bmatrix} \begin{bmatrix} x^* \\ v^* \end{bmatrix} = \begin{bmatrix} -q \\ b \end{bmatrix}$$

This is known as the $\textbf{KKT System}$ for the ECM and it forms the basis for how we will solve this problem using an extension of Newton's Method. The coefficient matrix is known as the $\textbf{KKT Matrix}$

$\textbf{Reduction}$

It is worth noting that we can reduce the ECM problem by eliminating the equality constraints and then solving the problem via unconstrained methods. First we paramaterize the constraint by finding a matrix $F \in \mathbb{R}^{nx(n-p)}$ and a vector $\hat{x} \in \mathbb{R}^n$ such that:

$$\{x|Ax = b \} = \{Fz + \hat{x} | z \in \mathbb{R}^{(n-p)}$$

Where $\hat{x}$ is any feasible point. We have effectively turned the ECM problem into the unconstrained problem

$$\text{minimize } \tilde{f}(z) = f(Fz + \hat{x})$$

$\textbf{The Newton Step}$

A good way we have to solve these problems is called $\textbf{Newton's Method with Equality Constraints}$ which is an iterative method stemming from the Taylor approximation of the objective $f$.

To derive this algorithm, we first substitute the second-order Taylor approximation of $f$ in the neighborhood of some feasible point $x$. Doing this we obtain the following problem which is approximately equivalent to the original:

$$\text{minimize }\hat{f}(x + v) = f(x) + \nabla f(x)^T v + \frac12 v^T \nabla^2 f(x) v$$
$$\text{subject to }A(x + v) = b$$

Our Newton Step, denoted $\Delta x_{nt}$, will be the optimal solution to this problem. Another way to think about this is $x_{nt}$ is the solution difference between the original problem and the above problem. We assume that the KKT matrix of this problem is nonsingular. Otherwise, we will not have a solution.

Using the same steps that we used in the quadratic problem to find the KKT system, we see that we can refactor this problem to form the following system:

$$\begin{bmatrix} \nabla^2 f(x) & A^T \\ A & 0 \end{bmatrix} \begin{bmatrix} \Delta x_{nt} \\ w \end{bmatrix} = \begin{bmatrix} -\nabla f(x) \\ 0 \end{bmatrix}$$

Notice that this is essentially the same system as in the quadratic problem except instead of $x^*$ we are solving for the Newton Step $\Delta x_{nt}$.

It is worth noting that the Newton step is always a feasible descent direction unless $x = x^*$

$\textbf{The Newton Decrement}$

Similar to the unconstrained minimization problem, the Newton decement will be the norm of the Newton Step:

$$\lambda(x) = \sqrt{\Delta {x_{nt}}^T \nabla^2 f(x) \Delta x_{nt}}$$

The similarities continue however. For $\hat{f}(x + v)$ being the second-order Taylor approximation of the objective, the distance between objective $f$ and the optimal solution to the Taylor approximation will be $\frac{\lambda(x)^2}{2}$ just as in the unconstrained optimization case. Thus, we can use this as our stopping criterion when computing $x^*$ numerically.

The algorithm to carry out this method is fairly straightforward. We start with a feasible $x$ satisfying $Ax = b$ and use that to compute $\Delta x_{nt}$ and $\lambda(x)$. We then iterate over $x_{n+1} = x_n + t \Delta x_{nt}$ while $\frac{\lambda^2}{2} > \epsilon$. We pick $t$ each iteration using backtracking line search just as in the unconstrained problem. 


This is a feasible descent method since if the starting iterate is feasible, all subsequent ones will be as well. Further we will have $f(x^{(k+1)}) < f(x^{(k)})$

$\textbf{An Example}$

To demonstrate this method, I will build a solver in Julia to solve ECM problems. Let us start with a simple example:

$$\text{minimize } x^2 - \sin(xy)$$
$$\text{subject to } \begin{bmatrix} 2 & 1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = 3$$

Below is a solver I wrote for this purpose. Its intent is to be easy to read and it is not necessarily the fastest nor most robust solution. Note that while this problem not convex, if we start in the same quadratic region, we are still able to solve it.

```julia
using LinearAlgebra
using Calculus

A = [2 1]
f(x,y) = x^2 - sin(x*y)
∇f(x,y) = (2x - y*cos(x*y), -cos(y*x)*x)
∇2f(x,y) = [sin(x*y)*y^2 + 2 x*y*sin(x*y) - cos(x*y); x*y*sin(x*y) - cos(x*y) x^2*sin(x*y)]
α = 0.49
β = 0.99
ϵ = 1e-12
t = 1
λ = 100 #prime λ with a big number
i = 0
x = [0 3] #starting point
xPlot = Float64[]
yPlot = Float64[]


while λ^2 / 2 > ϵ
    print("\n\nIteration ", i)
    i += 1
    
    #Compute Newton Step (p. 526 (10.11))
    J = [∇2f(x[1],x[2])[1,1] ∇2f(x[1],x[2])[1,2] 2; 
        ∇2f(x[1],x[2])[2,1] ∇2f(x[1],x[2])[2,2] 1; 
        2 1 0]
    c = -[∇f(x[1],x[2])[1]; ∇f(x[1],x[2])[2]; 0]
    v = J \ c
    Δxn = [v[1]; v[2]] #This is our Newton Step
    print("\nNewton step = ", Δxn)
    print("\nw =           ", v[3])

    #Compute Newton Decrement (p. 527 (10.12))
    λ = sqrt(transpose(Δxn) * ∇2f(x[1],x[2]) * Δxn)
    print("\nlamdba =      ",λ)

    #Choose t with backtracking line search (p. 464 alg. 9.2)
    t = 1
    while f(x[1] + t*Δxn[1], x[2] + t*Δxn[2]) > f(x[1], x[2]) + (α*t*[∇f(x[1], x[2])[1] ∇f(x[1], x[2])[2]]*Δxn)[1] 
        t *= β
    end

    #Update x
    x += t*transpose(Δxn)
    obj = f(x[1], x[2])
    
    #Store iteration data to plot later
    push!(xPlot,x[1])
    push!(yPlot,x[2])
    

    print("\n[x,y] =       ", x)
    print("\nObjective =   ", obj)

end

```

```julia
using Plots

_x = -2.5:0.025:2.5
_y = -10:0.025:10
con(x) = 3 - 2x
X = repeat(reshape(_x, 1, :), length(_y), 1)
Y = repeat(_y, 1, length(_x))
Z = map(f, X, Y)
p1 = Plots.heatmap(_x,_y, f, fill = true)

plot(p1)
plot!(con, color = "green", label = "Constraint") #feasible set
Plots.scatter!(xPlot, yPlot, color = "white", label = "Solution")
```

$\textbf{Making the Solver More Robust}$

Currently, our solver can only handle a 3-dimensional objective. We want to be able to handle any ECM of n-dimensions. The concepts remain the same but we are going to need to do some extra programming to automate problem construction.

```julia
using Calculus
using LinearAlgebra


#=
This function builds the coefficient matrix
in the system on page 526 (10.11) that is
used to compute the Newton Step
=#
function Build_J_Matrix(_A, _∇2f)
    a = size(_A,1)
    _At = transpose(_A)

    return [_∇2f _At; _A zeros(a,a)] 
end

#=
Computes the gradient _∇f at point _x
=#
function Compute_Gradient(_f, _x)
    return Calculus.gradient(g -> _f(g),_x)
end

#=
Computes the hessian _∇2f at point _x
=#
function Compute_Hessian(_f, _x)
    return hessian(h -> _f(h),_x)
end

#=
Builds c vector for Newton Step equation
=#
function Build_c(_∇f,_A, a, b)
    return [-_∇f; zeros(a)]
end

#=
Extracts Newton Step gradient from the
KKT equation
=#
function Extract_Step(_v, b)
    return _v[1:b]
end

#=
Computes Newton Step (p. 526 (10.11))
=#
function Compute_Newton_Step(_f, _∇f, _∇2f, _A, _x)
    #Compute Newton Step (p. 526 (10.11))
    a = size(_A,1)
    b = size(_∇f,1)
    _J = Build_J_Matrix(_A, _∇2f)
    _c = Build_c(_∇f, _A, a, b)
    _v = _J \ _c
    Δxn = Extract_Step(_v,b) #This is our Newton Step

    return Δxn
end

#=
Compute Newton Decrement (p. 527 (10.12))
=#
function Compute_Newton_Decrement(_Δxn, _∇2f)
    return abs(transpose(_Δxn) * _∇2f * _Δxn)
end

#=
Choose t with backtracking line search (p. 464 alg. 9.2)
=#
function Back_Line_Search(_x, _Δxn, _t, _f, fx, _∇f, _α, _β, _n)

    _t = 1
    
    while _f(_x + _t*_Δxn) > fx + _α*_t*transpose(_∇f)*_Δxn
        _t *= β
    end

    return _t
end


#=
Full solver: Minimize _g subject to _Ax = _b with tolerance = _ϵ
=#
function Solve_ECM(_g, _A, _b, _α, _β, _ϵ, x)

    obj = Float64
    i = 0
    t = 0.01
    λ2 = 10 #prime λ with a big number to ensure first iteration is run
    
    _n = length(x)
    it = 0

    while λ2 / 2 > ϵ
        
        #store plot data if 3d objective
        if _n == 2
            push!(xPlot, x[1])
            push!(yPlot, x[2]) 
        end
        
        obj = _g(x)
        
        print("\n\nIteration ", i)
        i += 1

        ∇g = Compute_Gradient(_g, x)
        ∇2g = Compute_Hessian(_g, x)
        Δxn = Compute_Newton_Step(_g, ∇g, ∇2g, _A, x)

        λ2 = Compute_Newton_Decrement(Δxn, ∇2g) #returns λ^2
        print("\nλ^2 =         ",λ2)

        t = Back_Line_Search(x, Δxn, t, _g, obj, ∇g, _α, _β, _n)

        #Update x
        x += t*Δxn

        print("\nObjective =   ", obj)
        it+=1
    end
    
    return obj, it
end
```

$\textbf{Interpreting the Results}$

So our solver indeed found the correct solution. However, it is extremely sensitive to the starting point. In this example I am using A/b as the starting point which is the least squares solution to the underdetermined system $Ax = b$. However, this has nothing to do with the objective so its solution could be very far away from the solution to the ECM. In order for this to be a viable solver, it would need a way to compute a good starting point.

$\textbf{Infeasible Start Newton Method}$

The Feasible Start Newton Method implemented above works very well when it is easy to find a feasible point. However, this is not always the case. For example, consider an objective function $f | \mathbb{R^n} \rightarrow \mathbb{R}$ with $dom f \neq \mathbb{R^n}$. Depending on the size of $n$ it can be very difficult to find a feasible point at all.

This is the main advantage to the $\textbf{Infeasible Start Newton Method}$ which operates similarly to its counterpart except that $x_0$ need not be feasible. For convenience, I will restate the optimality condition for ECMs.

To perform the Infeasible Start Newton Method, we start with a point $x_0$ which need not be feasible but does need to be in $dom f$. Our goal is to find a step $\Delta x$ that when applied to $x_0$ will yield a feasible point (or at least approximately so).

Similarly to the FSNM, we can evaluate the Taylor Approximation of the gradient at this point which we see is:

$$\nabla f(x + \Delta x) \approx \nabla f(x) + \nabla^2 f(x) \Delta x$$

Applying the equality constraint at our point $x + \Delta x$, we see the following:

$$\nabla f(x) + \nabla^2 f(x) \Delta x + A^T w = 0$$

Refactoring into matrix form yields:

$$\begin{bmatrix} \nabla^2 f(x) & A^T \\ A & 0 \end{bmatrix} \begin{bmatrix} \Delta x_{nt} \\ w \end{bmatrix} = -\begin{bmatrix} \nabla f(x) \\ Ax - b \end{bmatrix}$$

Now, we need the residuals which will be part of our stopping criterion later. For this we use the same optimality conditions as before:

$$Ax^* = b \text{ (primal condition)}$$
$$\nabla f(x^*) + A^T v^* = 0 \text{ (dual condition)}$$

Thus, the residual of the primal is $r_p(x,v) = Ax - b$ and the residual of the dual problem is $r_d(x,v) = \nabla f(x) + A^T v$.

Differentiating both and combining we get that the Taylor approximation of $r$ at $y$ is:

$$r(y + z) \approx \hat{r}(y + z) = r(y) + Dr(y)z$$

where $Dr(y)z$ is the derivative of $r$ at $y$. Thus, our primal-dual Newton step $\Delta y_{pd}$ is defined as:

$$Dr(y) \Delta y_{pd} = -r(y)$$

The book goes into more detail on page 532.

Combining all this with the system above gives us the equation to obtain the primal-dual step $x_{pd}$

$$\begin{bmatrix} \nabla^2 f(x) & A^T \\ A & 0 \end{bmatrix} \begin{bmatrix} \Delta x_{pd} \\ v^+ \end{bmatrix} = -\begin{bmatrix} \nabla f(x) \\ Ax - b \end{bmatrix}$$

Notice that so far, this is very similar to the last method. One major difference however is that $\Delta_{nt}$ in this case is not always a descent direction. To see this we differentiate $f(x + t\Delta x)$ with respect to $t$ and after some intermediate steps, we obtain:

$$\frac{d}{dt} f(x + t\Delta x) = -\Delta x^T \nabla^2 f(x) \Delta x + (Ax - b)^T w$$
$$\text{ full steps in the book p. 533}$$

Note that this is not necessarily negative unless $Ax = b$ meaning that this is not necessarily a descent direction. I implement this algorithm programmatically in the numerical expriments notebook.

$\textbf{The Stopping Criterion}$

In the FSNM, we stopped the algorithm when $\frac{\lambda^2}{2} < \epsilon$ for some tolerance $\epsilon$. In the ISNM, we use the norm of the residual as our stopping point. This is defined as the derivative of the residual norm evaluated at our current iterative approximation of $x^*$. Mathematically, we say this as:

$$-\|r(y)\|_2 = \frac{d}{dt} \|r(x + t \Delta y_{pd} \|_2$$

$\textbf{The Full ISNM Algorithm (p. 534)}$

To carry out this algorithm, start with an initial point $x_0 \in dom f$ which is not necessarily feasible. Then we compute the primal-dual Newton Step $\Delta y_{pd} = (\Delta x_{nt}, \Delta v_{nt})$. Now, pick $t$ using backtracking line search. The difference here is that we use $\|r\|_2$ instead of function evaluations as the stoping condition for the $t := \beta t$ step. 

Finally, we update $x := x + t \Delta x_{nt}$ and $v := v + t \Delta v_{nt}$. 

Continue in this way until $x$ is feasible and $\|r(x,v)\|_2 \leq \epsilon$ for some tolerance $\epsilon$.

$\textbf{Newton's Method For the Dual Problem}$



We will now look at solving the Lagrangian dual problem using Newton's method. We will consider the following problem with convex, quadratic objective:

$$\text{minimize } f(x) = \frac12 x^TPx + q^T x + r$$
$$\text{subject to } Ax = b$$

From this, we get the Lagrangian which is:

$$\mathcal{L}(x,v) =\frac12 x^TPx + q^T x + r + v^T(b - Ax)$$

Which we use to construct the optimality condition as before by taking the gradient with respect to $x$:

$$Px + q + A^T v = 0$$

Thus we see that $x^* = P^{-1}(-q - A^T v)$ and we can substutue back into our Lagrangian dual to obtain:

$$\mathcal{L}(x^*, v) = \frac12 {x^*}^TPx^* + q^T x^* + r + v^T (b - Ax^*)$$

This we can apply Newton's method to in the unconstrained form since the primal only has equality constraints. We are looking to minimize the negative of this objective. This will give us $v^*$ which in turn directly gives us $x^*$ since $x^* = P^{-1}(-q - A^T v)$.

To demonstrate this we will go through a simple example. Consider the following problem with quadratic objective:

$$\text{minimize } \frac12 \begin{bmatrix} x & y \end{bmatrix} \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix}$$
$$\text{subject to } \begin{bmatrix} 1 & -2 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = 5$$

Following the above procedure, we see that $x^* = \begin{bmatrix} 4v \\ -2v \end{bmatrix}$ which gives $\mathcal{L} (x^*,v) = -16v^2 + 5v$. The negative of this becomes our new unconstrained objective we want to minimize. Notice that the size of $v$ is the same as the number of equality constraints. This is true in general and makes solving the dual a very attractive alternative to solving the primal if the constraint count is relatively low.

```julia
using Printf
using Plots

function print_data(itterationCount, x, val)
   @printf("\ni = %1.0f | v: %20.20f | f(x): %20.20f", itterationCount - 1, x, val) 
end

function newton_method(fprime, fDprime, init, step, it, tol)
    print("\n\nNewton's Method | Initial Guess = ", init,"\n------------------------------")
    
    x = init
    t = 1
    
    _x = Float64[]
    _y = Float64[]
        

    
    #do first itteration
    x -= fprime(x)/fDprime(x)
        
    val = fprime(x) #prevents multiple function evaluations per iteration which would be a waste

    
    #Newton's method
  #  while val > tol
        print_data(i, x, f(x))
        x -= t*val/fDprime(x) 
        val = fprime(x)
        push!(_x,x)
        push!(_y,val)
  #  end
    
    return _x, _y
end


initialGuess = 1.0
f(x) = 16x^2 - 5x
fp(x) = 8x - 5
fpp(x) = 8

plot(f, -3,12, label = "L")
plot!(fp, -3,12, label = "Lprime")
tuple = newton_method(fp, fpp, initialGuess,1,6,1e-10)

plot!(tuple[1], tuple[2], label = "Iterations ",  seriestype = :scatter)



```

This simple, quadratic example converges exactly on the first iteration to give us $v = 0.625$. Which means $x^* = \begin{bmatrix} 4(0.625) \\ -2(0.625) \end{bmatrix} = \begin{bmatrix} 2.5 \\ -1.25 \end{bmatrix}$. This result is confirmed below.

```julia
using JuMP
using Ipopt

m5 = Model(Ipopt.Optimizer)

@variable(m5, -100 <= x <= 100)
@variable(m5, -100 <= y <= 100)

@NLconstraint(m5, c1, x - 2y == 5)

@NLobjective(m5, Min, x^2/2 + 2.5*x*y + 2*y^2)

JuMP.optimize!(m5)

print(value(x)," | ", value(y))
```
